```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("cases_by_activity_2026-05-13T11_30_06.719551Z.csv")

# View dataset shape
print(df.shape)

# View column names
print(df.columns)

# View first 5 rows
print(df.head())
```

```python

# Create CASE_TYPE column
df["CASE_TYPE"] = df["Case Number"].str.extract(r'^([A-Z]+)')

# View results
print(df[["Case Number", "CASE_TYPE"]].head())
```

```python
r'^([A-Z]+)' #means:

^ → start of the text
[A-Z]+ → capture all capital letters appearing first

So it stops before the /.#
```

```python
#Standardize column names
df.columns = [col.strip().upper() for col in df.columns]

# Count records by CASE_TYPE
case_type_counts = df['CASE_TYPE'].value_counts().reset_index()

# Rename columns
case_type_counts.columns = ['CASE_TYPE', 'COUNT']

# Display the table
print(case_type_counts)
```

```python
# Define relevant columns
case_type_col = "CASE_TYPE"
action_col = "COURT ACTION NAME"
```

```python
# Filter rows containing either Judgment or Ruling
filtered = df[
    df[action_col]
    .astype(str)
    .str.contains("Judgment|Ruling", case=False, na=False)
].copy()
```

```python
# Create a new column classifying the action
filtered["DECISION_TYPE"] = filtered[action_col].apply(
    lambda x: "Judgment"
    if "judgment" in str(x).lower()
    else "Ruling"
)
```

```python
# Create summary table
summary_table = pd.pivot_table(
    filtered,
    index=case_type_col,
    columns="DECISION_TYPE",
    aggfunc="size",
    fill_value=0
).reset_index()
```

```python
# Display result
print(summary_table)

# Optional: Save to CSV
summary_table.to_csv("judgment_ruling_by_case_type.csv", index=False)
```
